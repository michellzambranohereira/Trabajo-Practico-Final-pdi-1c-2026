```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\usuario\Trabajo-Practico-Final-pdi-1c-2026\ocr\imagenes\imagen_prueba.jpeg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray,(5,5),0)

_, thresh = cv2.threshold(
    blur,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

fig, axes = plt.subplots(1,4, figsize=(15,5))

axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original")

axes[1].imshow(gray, cmap="gray")
axes[1].set_title("Grises")

axes[2].imshow(blur, cmap="gray")
axes[2].set_title("Suavizado")

axes[3].imshow(thresh, cmap="gray")
axes[3].set_title("Binarizada")

for ax in axes:
    ax.axis("off")

plt.show()
```
